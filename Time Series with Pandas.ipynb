# Solution

**Imports**

```python
import datetime as dt
import pandas as pd
```

```python
df = pd.read_csv('Data/london_weather_MODIFIED - london_weather_MODIFIED.csv')
df.info()
df.head()
```

## 1.Convert the date column to datetime dtype. Double-check the result. Is it what you expected? If not, be sure to correct it by entering the appropriate format!

```python
# attempt to transform date to a datetime dtype
pd.to_datetime(df['date'])
```

- Pandas interpreted the int as Unix times and used it as the milliseconds after the 12:00 AM 01/01/1970

```python
# change to datetime with correct format
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.info()
df.head()
```

```python
# set index as date
df = df.set_index('date')
df.head()
```

## 2.Set the date column as the index. Inspect the new index of your dataframe. What type of index is it? Does it have a frequency?

```python
df.index
```

- It is a datetime index.
- It does not have a frequency.

## 3.Filter out only the required data:

- Keep only data from the year 2000 or later.
- Keep only the following features:
    - precipitation
    - mean_temp
    - min_temp
    - max_temp
    - snow_depth

```python
df_filtered = df.loc['2000':, ['precipitation',
                      'mean_temp', 
                      'min_temp', 
                      'max_temp', 
                      'snow_depth']]
df_filtered
```

## 4.Answer the following questions using .idxmax() and .idxmin(), print statements, and the .strftime() method. Use f-strings to print the following statements (with the blanks filled in):

### a) "The lowest temperature of {max temp} degrees occurred on {date as "MonthName 2-digit-day, 4-digit year. } which was a {day of the week}."

```python
min_temp_date = df_filtered['max_temp'].idxmin()
min_temp = df_filtered['max_temp'].min()

print(f"The lowest temperature of {min_temp} degrees occurred on {min_temp_date.strftime('%B %d, %Y')}, which was a {min_temp_date.day_name()}.")
```

### b) "The highest daily precipitation of {max preciptation} inches occurred on {date as "MonthName 2-digit-day, 4-digit year. } which was a {day of the week}."

```python
max_precip_date = df_filtered['precipitation'].idxmax()
max_precip = df_filtered['precipitation'].max()

print(f"The highest daily precipitation of {max_precip} inches occurred on {max_precip_date.strftime('%B %d,%Y')}, which was a {max_precip_date.day_name()}.")
```
